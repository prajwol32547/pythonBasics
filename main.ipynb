```python
x =[[1,2],[3,4]]
y =[[2,3],[3,4]]
z =[[0,0],[0,0]]
for k in range(0, len(x)):
    for l in range(0, len(y[0])):
      for m in range(0, len(y)):
         z[k][l] += x[k][m] * y[m][l]

print(z)
```
